```
def caesar_encrypt(text, shift):
  result = ""
  for char in text:
    if char.isalpha():
      shift_base = 65 if char.isupper() else 97
      result += chr((ord(char) - shift_base + shift) % 26 + shift_base)
    else:
      result += char
  return result

def caesar_decrypt(text, shift):
  return caesar_encrypt(text, -shift)


plain = "Anjay mabar"
cipher = caesar_encrypt(plain,2)
print("cipher", cipher)
print("decrypt:", caesar_decrypt(cipher,2))
```

```
def caesar_encrypt(text, shift):
  result = ""
  for char in text:
    if char.isalpha():
      shift_base = 65 if char.isupper() else 97
      result += chr((ord(char) - shift_base + shift) % 26 + shift_base)
    else:
      result += char
  return result

def caesar_decrypt(text, shift):
  return caesar_encrypt(text, -shift)


#inpur user
text = input ("masukan text: ")
shift = int(input("masukan shift: "))

cipher = caesar_encrypt(text,shift)
print("cipher:", cipher)
print("decrypt:", caesar_decrypt(cipher,shift))

#BRUTE FORCE
print("\nBrute Force Result")
for i in range(26):
    print(f"shift {i}: {caesar_decrypt(cipher, i)}")
```
